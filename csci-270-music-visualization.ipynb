```python
import librosa
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
def visualize_notes_of(filename):
    x, sr = librosa.load(filename)
    hop_length = 512
    chromagram = librosa.feature.chroma_stft(x, sr=sr, hop_length=hop_length)
    plt.figure(figsize=(15, 5))
    librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')
```

```python
visualize_notes_of('../input/csci-270-music-2022/Green Day - Good Riddance (Time of Your Life).mp3')
```

```python
visualize_notes_of('../input/csci-270-music-2022/Halo 2.mp3')
```

```python
visualize_notes_of('../input/csci-270-music-2022/Song_LOZ_WindWaker.mp3')
```
